```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read CSV data
# Replace 'your_data.csv' with your actual CSV file name
df = pd.read_excel("wind diagram.xlsx")  # tab-separated

# Convert wind direction from degrees to radians for polar plot
df['Wind direction (radians)'] = np.deg2rad(df['Wind direction'])

# Set up polar plot
plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)

# Plot using scatter (with PM10 as radius and Wind direction as angle)
sc = ax.scatter(
    df['Wind direction (radians)'],
    df['PM10'],
    c=df['Wind speed'],
    cmap='coolwarm',
    s=50 + df['Wind speed'] * 200,  # Size reflects wind speed
    alpha=0.75,
    edgecolors='black'
)

# Add colorbar
cbar = plt.colorbar(sc, ax=ax, orientation='vertical', label='Wind Speed (m/s)')

# Chart formatting
ax.set_theta_zero_location('N')  # 0° at the top
ax.set_theta_direction(-1)       # Clockwise
ax.set_title('Polar Chart of PM10 vs Wind Direction and Speed', fontsize=14)
ax.set_rlabel_position(135)      # Move radial labels

# Show plot
plt.tight_layout()
plt.show()
```
